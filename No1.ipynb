```python
import json

# Tariff data structure from the provided tables
data_bundle_tariffs = {
    "A": {"tariff": 0.8, "currency": "USD", "unit": "megabytes"},
    "B": {"tariff": 0.9, "currency": "USD", "unit": "megabytes"},
    "C": {"tariff": 5, "currency": "USD", "unit": "megabytes"},
}

voice_bundle_tariffs = {
    "A": {"tariff": 0.5, "currency": "USD", "unit": "minute"},
    "B": {"tariff": 0.2, "currency": "USD", "unit": "minute"},
    "C": {"tariff": 0.3, "currency": "USD", "unit": "minute"},
}

# taking the databunle data
def data_bundle_tariff(country):
    if country in data_bundle_tariffs:
        return json.dumps(data_bundle_tariffs[country])
    else:
        return json.dumps({"error": "Country not found"})


# Taking the Voice Bundle Data
def voice_bundle_tariff(country):
    if country in voice_bundle_tariffs:
        return json.dumps(voice_bundle_tariffs[country])
    else:
        return json.dumps({"error": "Country not found"})

# Example usage:
print(data_bundle_tariff("A"))   # {"tariff": 0.8, "currency": "USD", "unit": "megabytes"}
print(voice_bundle_tariff("B"))  # {"tariff": 0.2, "currency": "USD", "unit": "minute"}
print(data_bundle_tariff("C"))   # {"tariff": 5, "currency": "USD", "unit": "megabytes"}
print(voice_bundle_tariff("D"))  # {"error": "Country not found"}
```
